Question 1

Calculate MSE and RMSE for Zomato Delivery Times

```python
import numpy as np 
from sklearn.metrics import mean_squared_error 
actual_times = np.array([25, 30, 35, 40, 28, 32, 45, 38, 27, 36]) 
predicted_times = np.array([27, 29, 34, 42, 30, 31, 43, 40, 26, 35]) 
mse = mean_squared_error(actual_times, predicted_times) 
rmse = np.sqrt(mse) 

print("MSE:", mse) 

print("RMSE:", rmse)
```

Residual Plot for Movie Ratings

```python
import numpy as np 
import matplotlib.pyplot as plt 
actual = np.array([ 4.5,4.2,3.8,4.0,4.8, 3.5,4.1,4.3,3.9,4.6, 4.0,4.4,3.7,4.9,4.2, 3.8,4.5,4.1,4.7,3.9 ]) 
predicted = np.array([ 4.4,4.1,3.9,4.1,4.7, 3.6,4.0,4.2,4.0,4.5, 4.1,4.3,3.8,4.8,4.1, 3.9,4.4,4.0,4.6,4.0 ]) 
residuals = actual - predicted 
plt.scatter(predicted, residuals) 
plt.axhline(y=0) 
plt.xlabel("Predicted Ratings") 
plt.ylabel("Residuals") 
plt.title("Residual Plot") 
plt.show()
```

Question 5

Compare Two Regression Models

```python
import numpy as np 
from sklearn.metrics import ( mean_squared_error, mean_absolute_error, r2_score ) 
actual = np.array([50, 60, 70, 80, 90]) 
model1 = np.array([52, 58, 69, 82, 88]) 
model2 = np.array([50, 60, 70, 80, 90]) 
for name, prediction in [ ("Model 1", model1), ("Model 2", model2) ]: 
    mse = mean_squared_error(actual, prediction) 
    rmse = np.sqrt(mse) 
    mae = mean_absolute_error(actual, prediction) 
    r2 = r2_score(actual, prediction) 
    print(name) 
    print("MSE:", mse) 
    print("RMSE:", rmse) 
    print("MAE:", mae) 
    print("R²:", r2) 
    print()
```
