## 2. Classification of Machine Learning Scenarios

### 1. Netflix recommending movies based on your watch history

**Type:** Supervised Learning

**Explanation:** Netflix uses past user behavior and preferences to predict and recommend movies that the user is likely to watch.

### 2. Spotify grouping similar songs into playlists

**Type:** Unsupervised Learning

**Explanation:** Spotify identifies patterns and similarities among songs and groups them without predefined labels.

### 3. A self-driving car learning to park by trial and error

**Type:** Reinforcement Learning

**Explanation:** The car learns through rewards and penalties based on the success or failure of its parking actions.

## 4. Traditional Programming vs Machine Learning for a WhatsApp Spam Filter

In traditional programming, developers create fixed rules to identify spam messages. For example, if a message contains words such as "win money," "free prize," or includes many suspicious links, it is marked as spam. The system can only detect spam that matches the predefined rules.

In machine learning, the system is trained using a large dataset of spam and non-spam messages. Instead of relying on manually written rules, the model learns patterns from the data and can identify new spam messages that it has never seen before. This makes machine learning more flexible and effective for handling changing spam techniques.
