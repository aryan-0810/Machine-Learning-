```python
import pandas as pd
```

```python
data = { "player_name": ["Virat", "Rohit", "Bumrah", "Gill", "Hardik"], "player_age": [35, None, 30, None, 31], "team": ["RCB", "MI", None, "GT", None], "venue": ["Mumbai", "Delhi", "Mumbai", "Ahmedabad", "Delhi"], "player_role": ["Batsman", "Batsman", "Bowler", "Batsman", "Allrounder"] }
```

```python
df = pd.DataFrame(data)
df
```

```python
print(df.isnull().sum())
```

Question 2
Fill Missing Values in player_age Using Median

```python
median_age = df["player_age"].median()
df["player_age"] = df["player_age"].fillna(median_age)
print(df["player_age"])
```

Question 3
Replace Missing Team Names with 'Unknown'

```python
df["team"] = df["team"].fillna("Unknown")
print(df.head(10))
```

Question 4
Apply One-Hot Encoding on venue Column

```python
venue_encoded = pd.get_dummies(df, columns=["venue"]) 
print(venue_encoded)
```

Question 5
Encode player_role Using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 

df["player_role_encoded"] = le.fit_transform(df["player_role"]) 
print(df[["player_role", "player_role_encoded"]]) 

mapping = dict(zip(le.classes_, le.transform(le.classes_))) 

print(mapping)
```
