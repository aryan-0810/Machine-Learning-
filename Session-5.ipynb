```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
instagram_hours = np.array([1, 2, 3, 4, 5, 6, 7]) 
battery_percentage = np.array([85, 80, 75, 70, 65, 60, 55]) 
plt.scatter(instagram_hours, battery_percentage) 
plt.xlabel("Instagram Usage (Hours)") 
plt.ylabel("Battery Percentage") 
plt.title("Instagram Usage vs Battery Percentage") 
plt.show()
```

Question 2
Train a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression 
X = instagram_hours.reshape(-1, 1) 
y = battery_percentage 
model = LinearRegression() 
model.fit(X, y) 
print("Slope (Coefficient):", model.coef_[0]) 
print("Intercept:", model.intercept_)
```

Question 3
Plot the Regression Line

```python
predicted_values = model.predict(X) 
plt.scatter(instagram_hours, battery_percentage) 
plt.plot( instagram_hours, predicted_values ) 
plt.xlabel("Instagram Usage (Hours)") 
plt.ylabel("Battery Percentage") 
plt.title("Regression Line") 
plt.show()
```

Question 4
Calculate Mean Squared Error (MSE)

```python
from sklearn.metrics import mean_squared_error 
mse = mean_squared_error( battery_percentage, predicted_values ) 
print("Mean Squared Error:", mse)
```
