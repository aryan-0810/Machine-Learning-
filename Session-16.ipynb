Question 1: Calculate Accuracy, Precision, Recall and F1 Score

Given:

TP = 50
TN = 120
FP = 10
FN = 20
Calculations

```python
TP = 50
TN = 120
FP = 10
FN = 20
```

```python
Accuracy = (TP + TN) / (TP + TN + FP + FN)
Accuracy = (50 + 120) / (50 + 120 + 10 + 20)
Accuracy = 170 / 200
```

```python
Precision = TP / (TP + FP)
Precision = 50 / (50 + 10)
Precision = 50 / 60
```

```python
Recall = TP / (TP + FN)
Recall = 50 / (50 + 20)
Recall = 50 / 70
```

```python
F1_Score = 2 * (Precision * Recall) / (Precision + Recall)
F1_Score = 2 * (0.8333 * 0.7143) / (0.8333 + 0.7143)
```

Question 2: Function to Print a Confusion Matrix

```python
def print_confusion_matrix(tp, tn, fp, fn):

    print("Confusion Matrix")

    print("-----------------------")

    print("                 Predicted")

    print("               Positive  Negative")

    print(f"Actual Positive     {tp}        {fn}")

    print(f"Actual Negative     {fp}       {tn}")

print_confusion_matrix(
    tp=50,
    tn=120,
    fp=10,
    fn=20
)
```

Given:

TP = 60

Question 4: F1 Score for Food Delivery Fraud Detection

```python
tp = 40
fp = 20
fn = 10

precision = tp / (tp + fp)

recall = tp / (tp + fn)

f1_score = (
    2 * precision * recall
) / (
    precision + recall
)

print("F1 Score:", round(f1_score, 4))
```
