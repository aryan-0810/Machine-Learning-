| Point             | Supervised Learning                                                                                 | Unsupervised Learning                                            |
| ----------------- | --------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------- |
| Data              | Uses labeled data                                                                                   | Uses unlabeled data                                              |
| Goal              | Predict a known target or outcome                                                                   | Discover hidden patterns or groups                               |
| Learning          | Learns from input-output examples                                                                   | Finds relationships and similarities in data                     |
| Example           | Instagram feed ranking can learn from labeled user interactions such as likes, comments, and clicks | Spotify can group songs or users with similar listening behavior |
| Common Algorithms | Linear Regression, Decision Tree, Random Forest, SVM                                                | K-Means, Hierarchical Clustering, DBSCAN, PCA                    |
| Business Use      | Predict customer churn or classify spam messages                                                    | Segment customers based on their behavior                        |


```python
products = [
    {"product": "Product A", "price": 500, "reviews": 120},
    {"product": "Product B", "price": 550, "reviews": 150},
    {"product": "Product C", "price": 700, "reviews": 180},
    {"product": "Product D", "price": 5000, "reviews": 1200},
    {"product": "Product E", "price": 5500, "reviews": 1400},
    {"product": "Product F", "price": 6000, "reviews": 1600},
    {"product": "Product G", "price": 25000, "reviews": 80},
    {"product": "Product H", "price": 28000, "reviews": 100}
]

clusters = {
    "Cluster 1 - Budget Popular Products": ["Product A", "Product B", "Product C"],
    "Cluster 2 - Premium Popular Products": ["Product D", "Product E", "Product F"],
    "Cluster 3 - Luxury Less Popular Products": ["Product G", "Product H"]
}

print("Product Dataset:")
for product in products:
    print(product)

print("\nProduct Clusters:")
for cluster, items in clusters.items():
    print(cluster)
    print(items)
```
