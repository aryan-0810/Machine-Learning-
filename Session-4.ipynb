Question 1
StandardScaler on IPL Dataset

```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler 

df = pd.DataFrame({ "runs": [550, 420, 610, 480, 390], "wickets": [5, 12, 2, 8, 15], "matches": [14, 14, 14, 14, 14], "strike_rate": [145.2, 132.5, 151.8, 140.1, 128.6] }) 
scaler = StandardScaler() 
scaled_data = scaler.fit_transform(df) 

scaled_df = pd.DataFrame(scaled_data, columns=df.columns) 

scaled_df.to_csv("ipl_scaled.csv", index=False) 

print(scaled_df)
```

Question 4
RFE Using DecisionTreeRegressor on Zomato Dataset

```python
from sklearn.tree import DecisionTreeRegressor 
from sklearn.feature_selection import RFE 
df = pd.DataFrame({ "rating": [4.5, 4.2, 3.9, 4.8, 4.1], "votes": [500, 300, 200, 800, 250], "delivery_time": [30, 40, 45, 25, 35], "distance": [2, 5, 7, 1, 4], "average_cost_for_two": [1200, 900, 700, 1500, 850] }) 
X = df.drop("average_cost_for_two", axis=1) 
y = df["average_cost_for_two"] 
model = DecisionTreeRegressor(random_state=42) 
rfe = RFE(model, n_features_to_select=3) 
rfe.fit(X, y) 
print("Selected Features") 
for feature, selected in zip(X.columns, rfe.support_): 
    if selected: 
        print(feature) 
print("\nRanking") 
print(dict(zip(X.columns, rfe.ranking_)))
```

Question 5
Detecting and Removing Outliers Using IQR Method

```python
df = pd.DataFrame({ "user_ratings": [4.2, 4.5, 4.1, 4.6, 4.3, 4.4, 9.8, 1.0] }) 
Q1 = df["user_ratings"].quantile(0.25) 
Q3 = df["user_ratings"].quantile(0.75) 
IQR = Q3 - Q1 
lower_bound = Q1 - 1.5 * IQR 
upper_bound = Q3 + 1.5 * IQR 
clean_df = df[ (df["user_ratings"] >= lower_bound) & (df["user_ratings"] <= upper_bound) ] 
rows_dropped = len(df) - len(clean_df) 
print("Rows Dropped:", rows_dropped) 
print(clean_df)
```
